```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo de audio
file_path = librosa.example('brahms')
audio_data, sr = librosa.load(file_path, sr=None)  # Carga con la tasa de muestreo original

# Calcular caractersticas
# Chroma STFT
chroma_stft = librosa.feature.chroma_stft(y=audio_data, sr=sr)
# RMS
rms = librosa.feature.rms(y=audio_data)
# Spectral Centroid
spec_cent = librosa.feature.spectral_centroid(y=audio_data, sr=sr)
# Spectral Bandwidth
spec_bw = librosa.feature.spectral_bandwidth(y=audio_data, sr=sr)
# Spectral Rolloff
rolloff = librosa.feature.spectral_rolloff(y=audio_data, sr=sr)
# Zero Crossing Rate
zcr = librosa.feature.zero_crossing_rate(y=audio_data)
# MFCC
mfcc = librosa.feature.mfcc(y=audio_data, sr=sr, n_mfcc=13)  # 13 es un nmero comn de coeficientes

# Visualizar algunas de las caractersticas
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 1)
plt.plot(np.mean(chroma_stft, axis=1))
plt.title('Chroma STFT')
plt.xlabel('Chroma Bins')
plt.ylabel('Mean Intensity')

plt.subplot(3, 2, 2)
plt.plot(rms[0])
plt.title('RMS Energy')
plt.xlabel('Frames')
plt.ylabel('Amplitude')

plt.subplot(3, 2, 3)
plt.plot(spec_cent[0])
plt.title('Spectral Centroid')
plt.xlabel('Frames')
plt.ylabel('Frequency (Hz)')

plt.subplot(3, 2, 4)
plt.plot(spec_bw[0])
plt.title('Spectral Bandwidth')
plt.xlabel('Frames')
plt.ylabel('Frequency (Hz)')

plt.subplot(3, 2, 5)
plt.plot(zcr[0])
plt.title('Zero Crossing Rate')
plt.xlabel('Frames')
plt.ylabel('Rate')

plt.tight_layout()
plt.show()
```
